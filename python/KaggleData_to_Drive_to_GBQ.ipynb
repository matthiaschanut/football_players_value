**Step 2 : Import csv files into BigQuery tables**

Lien tutuorial : https://medium.com/@syamsulkamal574/the-easy-way-to-extract-transform-and-load-csv-data-from-google-drive-to-bigquery-with-python-74aec2e5363d

```python
#give permission to access files from Google Drive and BigQuery

from google.colab import auth
auth.authenticate_user()

import pandas as pd

%cd /content/gdrive/MyDrive/Kaggle

#define project_id and dataset name
project_id = 'transfermarkt-391212'
dataset_name = 'kaggle_data'

#open file csv
df1 = pd.read_csv('game_events.csv')
df2 = pd.read_csv('player_valuations.csv')
df3 = pd.read_csv('game_lineups.csv')
df4 = pd.read_csv('games.csv')
df5 = pd.read_csv('players.csv')
df6 = pd.read_csv('club_games.csv')
df7 = pd.read_csv('competitions.csv')
df8 = pd.read_csv('clubs.csv')
df9 = pd.read_csv('appearances.csv')

df1.to_gbq(destination_table=f'{dataset_name}.game_events', project_id=project_id, chunksize=None, if_exists='replace')
df2.to_gbq(destination_table=f'{dataset_name}.player_valuations', project_id=project_id, chunksize=None, if_exists='replace')
df3.to_gbq(destination_table=f'{dataset_name}.game_lineups', project_id=project_id, chunksize=None, if_exists='replace')
df4.to_gbq(destination_table=f'{dataset_name}.games', project_id=project_id, chunksize=None, if_exists='replace')
df5.to_gbq(destination_table=f'{dataset_name}.players', project_id=project_id, chunksize=None, if_exists='replace')
df6.to_gbq(destination_table=f'{dataset_name}.club_games', project_id=project_id, chunksize=None, if_exists='replace')
df7.to_gbq(destination_table=f'{dataset_name}.competitions', project_id=project_id, chunksize=None, if_exists='replace')
df8.to_gbq(destination_table=f'{dataset_name}.clubs', project_id=project_id, chunksize=None, if_exists='replace')
df9.to_gbq(destination_table=f'{dataset_name}.appearances', project_id=project_id, chunksize=None, if_exists='replace')
```
