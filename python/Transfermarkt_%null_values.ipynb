```python
from google.cloud import bigquery
import pandas as pd

# Set up Google Cloud credentials
from google.colab import auth
auth.authenticate_user()

# Connect to BigQuery
client = bigquery.Client(project='transfermarkt-391212')

# Define the dataset name
dataset_name = 'kaggle_data'

# Get a list of tables in the dataset
tables = client.list_tables(dataset_name)

# Function to calculate the percentage of null values in a DataFrame
def get_null_percentage(df):
    total_rows = df.shape[0]
    null_counts = df.isnull().sum()
    null_percentages = (null_counts / total_rows) * 100
    return null_percentages

# Loop through each table and calculate the percentage of null values for each column
for table in tables:
    table_name = table.table_id
    table_ref = client.dataset(dataset_name).table(table_name)
    table = client.get_table(table_ref)
    query = f"SELECT * FROM `{table_ref.project}.{table_ref.dataset_id}.{table_ref.table_id}`"
    df = client.query(query).to_dataframe()

    null_percentages = get_null_percentage(df)

    print(f"Table: {table_name}")
    print(null_percentages)
    print("\n")
```
